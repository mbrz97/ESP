```python
import numpy as np
from keras.callbacks import Callback

class NMSECallback(Callback):
    def __init__(self, x_val, y_val):
        self.x_val = x_val
        self.y_val = y_val

    def on_epoch_end(self, epoch, logs=None):
        predictions = self.model.predict(self.x_val)
        mse = np.sqrt(np.mean(np.square(predictions - self.y_val)))
        true_variance = np.sqrt(np.var(self.y_val))
        nmse = mse / true_variance
        print(f'\nEpoch {epoch + 1} - NMSE: {nmse:.4f}')

# Create the NMSE callback
nmse_callback = NMSECallback(x_val, y_val)
```

```python
from keras.utils import Sequence

class ESPDataGenerator(Sequence):
    def __init__(self, x_data, y_data, batch_size, shuffle=False, preprocess_function = None):
        self.x_data = x_data
        self.y_data = y_data
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indices = np.arange(len(self.x_data))
        self.preprocess_function = preprocess_function

        if self.shuffle:
            np.random.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.x_data) / self.batch_size))

    def __getitem__(self, index):
        start_idx = index * self.batch_size
        end_idx = (index + 1) * self.batch_size
        batch_indices = self.indices[start_idx:end_idx]
        #print(self.x_data.shape)
        batch_x = self.x_data[batch_indices]
        #print(batch_x.shape)
        batch_y = self.y_data[batch_indices]

        if self.preprocess_function is not None:
          batch_x = self.preprocess_function(batch_x)
          batch_y = self.preprocess_function(batch_y)

        # You can add any preprocessing steps here
        #print(batch_x.shape)
        return batch_x, batch_y

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

batch_size = 1

data_generator = ESPDataGenerator(X_train, y_train, batch_size)
#data_generator_val = ESPDataGenerator(x_val, y_val, batch_size)

epochs = 1

for epoch in range(epochs):
    for batch_x, batch_y in data_generator:
        # Train your model using the current batch
        pass
```
