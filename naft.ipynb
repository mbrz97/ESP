```python
def convert_strings_to_floats(data):
    if isinstance(data, list):
        return [convert_strings_to_floats(item) for item in data]
    elif isinstance(data, np.ndarray):
        return np.array([convert_strings_to_floats(item) for item in data])
    elif isinstance(data, str):
        try:
            return float(data)
        except ValueError:
            return data
    else:
        return data


x_data = convert_strings_to_floats(x_data)
y_data = convert_strings_to_floats(y_data)
```

```python
def remove_commas(data):
    if isinstance(data, list):
        return [remove_commas(item) for item in data]
    elif isinstance(data, np.ndarray):
        return np.array([remove_commas(item) for item in data])
    elif isinstance(data, str):
        return data.replace(',', '')
    else:
        return data

x_data = remove_commas(x_data)
y_data = remove_commas(y_data)
```

```python
from keras.utils import Sequence

class ESPDataGenerator(Sequence):
    def __init__(self, x_data, y_data, batch_size, shuffle=True, preprocess_function = None):
        self.x_data = x_data
        self.y_data = y_data
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indices = np.arange(len(self.x_data))
        self.preprocess_function = preprocess_function

        if self.shuffle:
            np.random.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.x_data) / self.batch_size))

    def __getitem__(self, index):
        start_idx = index * self.batch_size
        end_idx = (index + 1) * self.batch_size
        batch_indices = self.indices[start_idx:end_idx]
        #print(self.x_data.shape)
        batch_x = self.x_data[batch_indices]
        #print(batch_x.shape)
        batch_y = self.y_data[batch_indices]

        if self.preprocess_function is not None:
          batch_x = self.preprocess_function(batch_x)
          batch_y = self.preprocess_function(batch_y)

        # You can add any preprocessing steps here
        #print(batch_x.shape)
        return batch_x, batch_y

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

batch_size = 16

data_generator = ESPDataGenerator(x_train, y_train, batch_size)
data_generator_val = ESPDataGenerator(x_val, y_val, batch_size)

epochs = 2

for epoch in range(epochs):
    for batch_x, batch_y in data_generator:
        # Train your model using the current batch
        pass
```
